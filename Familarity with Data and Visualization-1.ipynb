# SQL

```python
import sqlite3

conn = sqlite3.connect('flower.db') 
c = conn.cursor()

c.execute('''
          CREATE TABLE IF NOT EXISTS flower
          ([quantity] INTEGER PRIMARY KEY, [name] TEXT, [price] INTEGER)
          ''')
          
c.execute('''
          INSERT INTO flower (quantity,name,price)

                VALUES
                (1,'jasmine',1800),
                (2,'rose',200),
                (3,'sunflower',300),
                (4,'lotus',450),
                (5,'lilly',150)
          ''')                     

conn.commit()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('flower.db') 
          
sql_query = pd.read_sql_query ('''
                               SELECT
                               *
                               FROM flower
                               ''', conn)

df = pd.DataFrame(sql_query, columns = ['quantity', 'name', 'price'])
df.head()
```

# MP4

```python
import cv2
import numpy as np
cap = cv2.VideoCapture('thiru.mp4')
if (cap.isOpened()== False):
    print("Error opening video file")
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == True:
            cv2.imshow('Frame', frame)
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
            else:
                    break
                    cap.release()
                    cv2.destroyAllWindows()
```
